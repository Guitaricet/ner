```python
import random

import pandas as pd
import numpy as np
```

```python
ALPHABET = ['9', '<UNK>', 's', 'а', '>', 'k', '[', 'p', '»', 'г', 'т', 'j', 'е', '0', 'q', '&', '•',
            '2', 'v', '%', '6', 'ь', 'd', 'z', 'r', '=', 'u', 'ч', '-', 'g', 'о', '–', ']', ',', 'ш',
            'i', 'ъ', ':', '©', '8', 'ж', 'к', 'н', '\xad', 'щ', 'c', '+', 't', 'ю', 'в', '№', 'w', '5',
            '#', 'y', '«', ')', 'л', 'n', '“', '3', 'ц', 'ы', '…', '(', 'e', 'h', 'x', 'х', '1', 'у', 'm',
            'l', 'и', 'з', '—', 'б', 'д', 'a', '_', 'ё', '$', 'р', '€', 'f', '<', 'э', 'o', '.', 'с', ';',
            '"', '?', '”', 'п', 'ф', '/', 'м', '|', 'й', '4', '*', "'", '<PAD>', 'b', '!', 'я']

def noise_generator(string, noise_level=0.1):
    assert isinstance(string, str)
    noised = ""
    for c in string:
        if random.random() > noise_level:
            noised += c
        else:
            noised += random.choice(ALPHABET)
    return noised

def read_data(datapath):
    with open(datapath) as f:
        xy_list = list()
        tokens = list()
        tags = list()
        for line in f:
            items = line.split()
            if len(items) > 1 and '-DOCSTART-' not in items[0]:
                token, tag = items
                if token[0].isdigit():
                    tokens.append('#')
                else:
                    # is you want static noise, add it here
                    tokens.append(token.lower())
                tags.append(tag)
            elif len(tokens) > 0:
                xy_list.append((tokens, tags,))
                tokens = list()
                tags = list()
    return xy_list
```
